```python
#Again, first thing to do is import Panda
import pandas as pd
```

```python
#Lets start with creating two dataframes with dictionaries.
df1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                    'B': ['B0', 'B1', 'B2', 'B3'],
                    'C': ['C0', 'C1', 'C2', 'C3'],
                    'D': ['D0', 'D1', 'D2', 'D3']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                    'B': ['B4', 'B5', 'B6', 'B7'],
                    'C': ['C4', 'C5', 'C6', 'C7'],
                    'D': ['D4', 'D5', 'D6', 'D7']},
                    index=[4, 5, 6, 7])
```

```python
df1
```

```python
df2
```

# Concatenation

```python
#We can concatenate those 2 dataframes vertically or horizontally based on axis. .concat()
dfmv = pd.concat([df1,df2],axis=0)
dfmv
```

```python
#Concatenating horizontally:
dfmh = pd.concat([df1,df2],axis=1)
dfmh
```

```python
#As you can see since the indecies are different, this created a lot of null information
#Filling that info with fillna()
dfmh.fillna(method="ffill", inplace=True)
```

```python
dfmh.fillna(method="bfill", inplace=True)
```

```python
dfmh
```

```python
#However, we can also fill in the previous dataframes while resetting the index, removing any nulls
pd.concat([df1,df2.reset_index(drop=True)],axis=1)
```
