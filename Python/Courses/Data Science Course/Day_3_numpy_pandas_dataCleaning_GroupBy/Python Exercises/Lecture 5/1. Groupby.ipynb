```python
#Import first thing
import pandas as pd
```

```python
#create a dataframe
data = {'Store':['Walmart','Walmart','Costco','Costco','Target','Target'],
       'Customer':['Tim','Jermy','Mark','Denice','Ray','Sam'],
       'Sales':[150,200,550,90,430,120]}

df = pd.DataFrame(data)
df
```

# Groupby

```python
#Now lets use groupby to group a specific column together. You need to set it in a variable
store = df.groupby("Store")
```

```python
#since we have grouped them, we can use aggitated on any integer column related to that specific group,
#This allows us to grab an idea of the relation between these groups and the data
#Since we have duplicate stores, it will find the mean on all duplicates and combine them on 1
store.mean()
```

```python
#We can also locate a specific store and its mean value, used for large datasets
store.mean().loc["Target"]
```

```python
#since we have grouped stores, we can now find all useful information for each store using .describe()
#helps in comparison of data instead of finding the total statisics of all data even if unrelated
store.describe()
```

```python
#and in case you dont like that style, you can still use the transpose style
store.describe().transpose()
```

```python
#This also helps in calling a specific column
store.describe().transpose()["Target"]
```

# Lets try Manipulating null variables based on the group

```python
#This is me adding extra rows with null information
df.loc["6"] = ["Costco", "Mark", None]
df.loc["7"] = ["Target", "Twayne", None]
df.loc["8"] = ["Walmart", "Crook", None]
df
```

### What if we want to fill those nulls with means? Lets do it

```python
#First we need to group them again to apply the new rows
store = df.groupby("Store")
#Now, we need to replace all Null with the mean based on the store. 
df["Sales"] = store["Sales"].transform(lambda x : x.mean())
df
```
