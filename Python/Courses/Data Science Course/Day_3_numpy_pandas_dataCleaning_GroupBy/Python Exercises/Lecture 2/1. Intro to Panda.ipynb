# Pandas Indexing

```python
#you can change pandas indexing with your own indexes. Doing so will help create unique indexes
new_index_values = ['A','B','C','D','E','F','G','H','I','J','K','L','M','N','O','P','Q']
drug.index=new_index_values
```

```python
drug.head()
```

```python
#You can also create subsets with loc or iloc. ORDER MATTERS
#iloc uses indexing to create the database
subset = drug.iloc[[1,2,3,4,5], [4,5]]
subset
```

```python
#loc uses names of indexes as well as columns to create subsets
subset = drug.loc[['B','C','D','E','F'], ['marijuana-use','marijuana-frequency']]
subset
```

```python
#You can also use the data from a column to set indexes.
drug.index = drug['age']
drug.head()
```

```python
#if you dont like the indexes youve set, you can reset them.
drug.reset_index(drop=True, inplace=True)
drug.head()
```

# Filtering Logic

```python
#You can also filter your data depending on conditions you have set.
condition1 = drug["marijuana-use"] > 20
condition2 = drug["cocaine-use"] > 2
drug[condition1 & condition2]
```

```python
#another way to write it
filters = drug[(drug["marijuana-use"] > 20) & (drug["cocaine-use"] > 2)]
```

```python
#Filtering Data and only displaying certain columns
subset = filters.iloc[list(range(0,7)),[3,6]]
subset
```
