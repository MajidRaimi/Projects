**Moving forward, let's add `age` and `name` as the class `properties/attributes`.** 

````python
# A simple class
class Duck:
    """
    This is a class Duck
    It has two methods quack and walk
    name = ""
    age = 0
    """
    
    def __init__(self, name = "default", age=0): # Note the defaults.
        self.name = name # All pets have a name.
        self.age = age # All pets have an age.
        print(name, "created")
    
    def quack(self):
        print('Quaaack! I am called', self.name,
              "; I am ", self.age, "month old")

    def walk(self):
        print(self.name, 'Walks like a duck.')
        
    # I am adding a static method here
    @staticmethod
    def info():
        print('This is a static method')
````

Now, in the class Duck!<br>
Each Duck has:
* name
* age
* can quack (class method)
* can walk (class method)

## Class Inheritance

#### Discussion: Similar Classes
* Phone is a class — there are hundreds of types of phones. What attributes and functions would a Phone have?
* What about an iPhone? Or android_phone?

***iPhone and android_phone would be objects of the Phone class.
But, there are different types of iPhones and Android phones.***<br>
>Should `IPhone` and `AndroidPhone` be classes themselves?
**What would you do?**

#### Introduction: Inheritance
* AndroidPhone and IPhone are separate classes and in the Phone class. ***This is called inheritance: making classes that are subsets of other classes.***
* Phone is the parent class. It’s a regular class! All phones:
    * Have a phone number. 
    * Can place phone calls. 
    * Can send text messages.

* IPhone is a child class. The child class inherits methods and properties from the parent class but can also define its own functionality. 
* Assume, iPhones uniquely (Well, now every phone have this!):
    * Have an unlock method that accepts a fingerprint.
    * Have a set_fingerprint method that accepts a fingerprint.

So, our assumption for the main class (parent) class is:
* All phones have a phone number
* Can place phone calls, and 
* Can send text messages.

Good!

````python
# Let's create a class "Phone"
# it should be a piece of cake now!
class Phone:
    def __init__(self, phone_number): 
        self.number = phone_number 
        
    def call(self, other_number):
        print("Calling from", self.number, "to", other_number)
        
    def text(self, other_number, msg):
        print("Sending text from", self.number, "to", other_number) 
        print("message is:", msg)
````

````python
phone = Phone(12345)
````

````python
phone.call(6789)
````

````python
phone.text(678, 'hello world')
````

````python
# till here, it should be fine, right?
````

### Underneath the `Phone class` definition, let’s create the `IPhone class`.

````python
# Class definitions accept a parameter specifying what class they inherit
class IPhone(Phone):
     def __init__(self, phone_number):
            super().__init__(phone_number) 
            
            # super()` calls the `init` defined in the parent class.
# Now self.number is set, 
# because that's what happens in the parent Phone class
````

````python
iphone = IPhone(234) 
# error without number -- see the parent class, needs phone_number
````

````python
# text() is from parent class
iphone.text(678, 'hello world')
````

````python
# call() is from parent class
iphone.call(567) # 234 in the instance above
````

**Now, let's add a new method, say unique to iPhone in the iPhone class.**

````python
class IPhone(Phone):
    """
    This is a child class for Phone class.
    This class inherit call and text functions from Phone class
    
    This class is specific to the iPhone
    """
    def __init__(self, phone_number):
            super().__init__(phone_number) 
            # Under the call to super, we can define unique IPhone variables. 
            # Regular Phone objects won't have this!
            
            self.fingerprint = None # we want it in multiple funcs below
            
    def set_fingerprint(self, finger_print):
        """This function sets the finger prints
        """
        self.fingerprint = finger_print
        print(self.fingerprint)
        
    def unlock(self, finger_print=None):
        """This function unlock the phone based on the 
        input.
        finger_print = None
        """
        if (self.fingerprint == None):
            print("Phone unlocked, no fingerprint needed!")
        elif (finger_print == self.fingerprint):
            print("Phone unlocked, Fingerprint matched!")
        else:
            print("Phone locked, Fingerprint did not match!")
        #self.fingerprint = fingerprint
        #print(self.fingerprint)
````

````python
# Let's set the fingerprints
iphone2.set_fingerprint('Qazi') # <shift+tab> for doc sting
# what is the value of self.fingerprint now?
````

### Overwriting attributes:
Let's start with another regular class `Building`.

````python
class Building(): 
    """No __init__ -- there are no instance variables to declare!
    we are only using class variables.
    This is possible in any class, not just inheritance. (Building is a normal class)
    """
    # Class variables/attributes 
    avg_sqft = 12500 
    avg_bedrooms = 3
    
    def describe_building(self): 
        print('Avg. Beds:', self.avg_bedrooms) 
        print('Avg. Sq. Ft.:', self.avg_sqft)
    
    def get_avg_price(self):
        price = self.avg_sqft * 5 + self.avg_bedrooms * 15000 
        return price
````

````python
my_building = Building() # <shift + tab>
````

````python
my_building.avg_bedrooms, my_building.avg_sqft
# press tab after b. and you will see all the class variables and methods!
````

````python
my_building.describe_building()
````

````python
my_building.get_avg_price()
````

**Let's inherit the class `Building` in a child class `Mansion`.**

```Python
class Mansion(Building):
    """
    Our child class definition goes here.
    We will have the same class variables, instance variables, 
    and methods as Mansion
    """
    ```

Well, <br>
* ***What if we want the class variables to have different values?***<br>
* Good thing is, ***we can set new ones.*** 

***Remember, child classes do not affect the parent class.***

````python
class Mansion(Building):
    # Overwrite the class variables. 
    avg_sqft = 6
    avg_bedrooms = 1
    # We don't have a call to super __init__. Why? 
    # There's no __init__ in the parent to call!
````

````python
### Now, let's try it out. ###
# This still has the old values. 
my_mansion = Mansion()
````

````python
my_mansion.get_avg_price()
# notice the cnage in the price, it is considering the overwritten values!
# get_avg_price() is from the parent class
````

````python
# Just to reconfirm from the parent class
bld = Building()
bld.get_avg_price()
````

## Summary and Q&A
#### Inheritance:
* Allows us to make classes using other classes as templates.
* Has a parent class (e.g. Phone) and a child class (e.g. IPhone). 
* The parent class is still a usable class!

#### Child classes:

* inherit methods and properties from a parent class. 
* Have access to all of the functionality of its parent.
* Can have new attributes and methods. They won’t be available to the parent.
* Can overwrite values from the parent class.
