```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
def plotClass(X,y,p):
    plt.figure()
    for i in range(y.shape[1]):
        if y[0,i]==0:
            plt.plot(X[0,i],X[1,i],'r'+p)
        else:
            plt.plot(X[0,i],X[1,i],'b'+p)
            
    plt.show()
```

```python
num_data=100 # data points per class

x1=np.random.randn(2,num_data)+4
x0=np.random.randn(2,num_data)

y1=np.ones((1,num_data))
y0=np.zeros((1,num_data))

X=np.concatenate((x1,x0),axis=1)
y=np.concatenate((y1,y0), axis=1)
plotClass(X,y,'o')
print(X.shape)
print(y.shape)
```

```python
def get_z(W, X, b):
  #w0+w1*x1+w2*x2
  return W@X+b


def get_sigmoid(z):
  return 1/(1+np.exp(-z))


def get_loss(y,yhat):
  return np.sum(-y*np.log(yhat) -(1-y)*np.log(1-yhat), axis=1)/ y.shape[1]


W = np.random.randn(1,X.shape[0])
b = 0
ls =[]
lr=0.0001
for i in range(10000):
  z = get_z(W,X,b)
  yhat = get_sigmoid(z)

  loss = get_loss(y,yhat)
  ls.append(loss)
  # dZ=dJ/dy * dy/ dZ
  dZ = yhat - y

  dW = dZ @ X.T
  db = np.sum(dZ)


  W -=lr*dW
  b -=lr*db


plt.plot(ls)
```

```python
plotClass(X, yhat>0.5,'o')
plotClass(X,y,'o')
```
