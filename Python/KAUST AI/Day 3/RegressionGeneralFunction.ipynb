```python
#Linear Regression for general functions
import numpy as np
import matplotlib.pyplot as plt
import math
pi=math.pi


x0=np.linspace(-1,1,100)

y0=np.sin(2*pi*.5*x0)+0.3*np.random.randn(x0.size)
plt.plot(x0,y0)
```

```python
# Create regression model
weightConstants = []

noOfBases = 3
for i in range(noOfBases + 1):
    weightConstants.append(x0 ** i)

weightConstants = np.array(weightConstants).T

weights = np.linalg.inv(weightConstants.T @ weightConstants) @ (weightConstants.T @ y0)

PredictedValues = weightConstants @ weights

plt.plot(x0,y0,x0,PredictedValues)
```
